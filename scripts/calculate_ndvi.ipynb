```python
import numpy as np

def compute(ins, outs):
    red = ins['Red16'].astype('float32') / 65535.0
    nir = ins['Nir16'].astype('float32') / 65535.0
    denom = nir + red
    ndvi = np.zeros_like(denom, dtype='float32')
    m = denom != 0
    ndvi[m] = (nir[m] - red[m]) / denom[m]
    ndvi = np.clip(ndvi, -1.0, 1.0)
    outs['NDVI'] = ndvi
    return True
```
